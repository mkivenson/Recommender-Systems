```python
skills = pd.read_csv("skills.csv", header = None, names = ["name"])
skills.head(n=10)
```

```python
input_skills = '''Software Engineer Full stack HTML CSS JavaScript Java Node.js 
PHP Python C++ Dart Kotlin C# Web Mobile AWS S3 API Gateway Lambda Voice Design BigQuery SQL PySpark Typescript Databases
Computer Science Robotics Reverse Proxy Server Relational Angular React React-Native Open Tracing Splunk Zipkin Vizceral'''
```

```python
# Import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

jobs_df["words"] = jobs_df["title"] + jobs_df["summary"] + jobs_df["description"]

# Create a TfidfVectorizer: tfidf
tfidf = TfidfVectorizer(stop_words="english", analyzer='word', vocabulary = map(lambda x: str(x).lower(), skills.name.values),
                            ngram_range=(1, 3), max_df=0.1, min_df=0.0025, max_features=None)

# Apply fit_transform to job descriptions: tfidf_jobs
job_summaries = jobs_df["words"].astype(str).copy()
job_summaries = job_summaries.append(pd.Series([input_skills]))  

tfidf_result = tfidf.fit_transform(job_summaries.values)

tfidf_jobs = pd.DataFrame(data = tfidf_result.toarray(), 
                          index = job_summaries, 
                          columns = tfidf.get_feature_names())

tfidf_jobs.tail()
```

```python
from sklearn.preprocessing import normalize
def get_recommendations(features, index):

    #cosine similarity
    norm_features = normalize(features.toarray())
    jobs_rec_df = pd.DataFrame(norm_features, index = index, columns = tfidf.get_feature_names())
    current_item = jobs_rec_df.loc[input_skills].values.reshape(-1,1)
    similarities = jobs_rec_df.dot(current_item).reset_index()
    recommended_jobs = similarities.nlargest(n=6, columns=0)
    recommended_jobs = pd.merge(recommended_jobs.reset_index(), 
                                jobs_df, 
                                left_on="index", 
                                right_on="words")[["title","company","summary","link"]]
    return jobs_rec_df, current_item, similarities, recommended_jobs
```

```python
jobs_rec_df, current_item, similarities, recommended_jobs = get_recommendations(tfidf_result, job_summaries)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
skills_rec = cosine_similarity(jobs_rec_df.transpose(), jobs_rec_df.transpose())
skills_rec_df = pd.DataFrame(skills_rec)
skills_rec_df.columns = list(skills.name.values)
skills_rec_df.index = list(skills.name.values)
skills_rec_df
```

```python
def skill_recommender(skill, matrix):
    norm_features = normalize(matrix.toarray())
    result = norm_features.sort_values(skill,ascending=False)
    print("Recommended Skill | Value")
    for i in range(1,5): 
        print (result[skill].index.values[i] + " | " + str(round(result[skill][i],3)))
    print("\n")
skill_recommender("Python", skills_rec)
```
